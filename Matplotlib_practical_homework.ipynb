### P4. Create the following pie chart.

````python
percentages = [31.3, 24.8, 12.4, 11.3, 10.8, 9.4]
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
cols = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62727', '#9367bd', '#8c564b']

plt.pie(percentages, 
        labels = languages, colors = cols, 
        startangle = 140, shadow = True, 
        explode = (0.2, 0, 0, 0, 0, 0), 
        autopct = "%1.1f%%", radius = 2,
        wedgeprops={"edgecolor":"k",'linewidth': 1},
        textprops = {"fontsize":11.5})
plt.title('Pie chart', y = 1.45, fontsize = 14)
plt.show()
````

### P6. Write a Python code to output the following graph. (or a similar one). 
```
The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
```

````python
from pandas import DataFrame 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
index = ['M','T','W','Th','F']

dic_data = {'Group1': g1, 'Group2': g2, 'Group3' : g3}

df = DataFrame(dic_data, index = index)
print(df)
ax = df.plot(kind = 'bar', title='Productivity during the week', rot = 0)
ax.set_xlabel('Groups')
ax.set_ylabel('Productivity')
````

### P7. Write a Python code to output the following graph. (or a similar one)
```
Draw 25 dots. Generate random x and y coordinates (integers from 1 to 30), random colors and random sizes (integers from 20 to 500) for the dots.
```

```
 25 .     x  y  (1- 30  int  ),      (20- 500  int  )
```

````python
import random

number_of_circles = 25 
coords = [i for i in range(0, 31, 5)]
x = [random.randrange(1, 31) for i in range(number_of_circles)] 
y = [random.randrange(1, 31) for i in range(number_of_circles)]
colors = np.array([np.random.randint(256, size = 3) for i in range(number_of_circles)]) 
area = [random.randrange(20, 501) for i in range(number_of_circles)]
plt.xticks(coords)
plt.yticks(coords)
plt.ylim(0, 30)
plt.xlim(0, 30)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Colorful dots")
plt.scatter(x, y, s = area, c = colors/255.0)

plt.show()
````

### P8. Plotly

```
Load the Iris dataset from plotly.express. 
```

````python
import plotly.express as px
df = px.data.iris()
df.head()
````

```
Create a scatter plot of petal_length vs petal_width. Use species as the color and display all the feature values when you hover over a point.
```

````python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)

fig = px.scatter(df, x="petal_length", y="petal_width", color="species", 
                size="sepal_length", 
                hover_data=['sepal_width'])
fig.show()
````

```
Create a histogram having 20 bins using the feature sepal_length. Use species feature as the color. 
```

````python
fig = px.histogram(df, x="sepal_length", color="species", nbins=20)
fig.show()
````

```
Plot sepal_length and sepal_width features on the same histogram. 
```

````python
import plotly.graph_objs as go
import plotly.figure_factory as ff
import plotly as py

x1 = df['sepal_length'] 
x2 = df['sepal_width']  
hist_data = [x1, x2]
group_labels = ['sepal_length', 'sepal_width']
fig = ff.create_distplot(hist_data, group_labels, bin_size=.3)
py.offline.iplot(fig)
````

```
Create a heatmap showing the correlation of the columns sepal_width and petal_width.
```

````python
import plotly.figure_factory as ff

my_df = (df.corr()).iloc[[1,3], [1, 3]]
z = my_df.values.tolist()
x = my_df.columns.tolist()
y = my_df.index.tolist()

fig = ff.create_annotated_heatmap(z, x=x, y=y, colorscale="Cividis_r")
fig.show()

````

```
Create a box plot showing the distribution of the sepal_length feature. 
```

````python
fig = px.box(df, x="sepal_length")
fig.show()
````
