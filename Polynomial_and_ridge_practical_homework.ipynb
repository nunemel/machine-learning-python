### Implement a function fit_ridge_regression

```python
def fit_ridge_regression(data_vector, response_vector, l=1):
    #(x_T*x+lI)-1* x_T*Y
    """
    :param data_vector: A pandas series object (i.e. a column of a dataframe),
                        where each element is a data point (x)
    :param response_vector: A pandas series object (i.e. a column of a dataframe),
                        where each element is label (y)
    :param l: ridge variable
    :return: numpy array beta of coefficients of the regression
    """
    #x
    data_vector = np.array(data_vector)
    #Y
    response_vector = np.array(response_vector)
    n = len(data_vector)
    k = 1 #independent feature count
    data_vector = data_vector.reshape((n, k))
    #add ones column for intercept
    data_vector = np.c_[np.ones(data_vector.shape[0]), data_vector]
    x_T = data_vector.T
    #identity matrix
    I = np.identity(2)
    dot_product = x_T.dot(data_vector) + l*I
    inverse = np.linalg.inv(dot_product)
    beta = inverse.dot(x_T.dot(response_vector)).flatten()
   
    return beta
```

### Implement a function fit_polynomial_regression

```python
def fit_polynomial_regression(data_vector, response_vector, degree=2):
    """
    :param data_vector: A pandas series object (i.e. a column of a dataframe),
                        where each element is a data point (x)
    :param response_vector: A pandas series object (i.e. a column of a dataframe),
                        where each element is label (y)
    :return: numpy array beta of coefficients of the regression
    """
    if ((degree) < 1): 
        pass
    
    data_vector = np.array(data_vector)
    response_vector = np.array(response_vector)
    n = len(data_vector)
    k = 1 #independent feature count
    data_vector = data_vector.reshape((n, k)).flatten()  
    data_vector = np.vander(data_vector, degree + 1, increasing=True) 
    x_T = data_vector.T
    dot_product = x_T.dot(data_vector)
    inverse = np.linalg.inv(dot_product)
    return inverse.dot(x_T.dot(response_vector))
    
    return beta
```

```python
def fit_ridge_polynomial_regression(data_vector, response_vector, degree=2, l=1.5):
    if ((degree) < 1): 
        pass
    #x
    data_vector = np.array(data_vector)
    #Y
    response_vector = np.array(response_vector)
    n = len(data_vector)
    k = 1 #independent feature count
    data_vector = data_vector.reshape((n, k)).flatten()   
    data_vector = np.vander(data_vector, degree + 1, increasing=True) 
    x_T = data_vector.T
    #identity matrix
    I = np.identity(degree + 1)
    dot_product = x_T.dot(data_vector) + l*I
    inverse = np.linalg.inv(dot_product)
    return inverse.dot(x_T.dot(response_vector))
```

```python
from sklearn.linear_model import RidgeCV

def getBestLambda(x, y, lambdas):
    ridgecv = RidgeCV(alphas = lambdas).fit(np.array(x).reshape(-1, 1),
                                          np.array(y).reshape(-1, 1))
    return ridgecv.alpha_.round(10)
```

### Calculate RMSE (root mean squared error) for ridge and polynomial regressions

```python
def rmse(my_beta, m_x, m_y):
    return np.sqrt(sum((y - sum(my_beta[i] * x ** i for i in range(len(my_beta)))) ** 2
            for x, y in zip(m_x, m_y)) / len(m_x))

def func_y(my_beta, m_x):
    curve_y = [sum(my_beta[i] * x ** i for i in range(len(my_beta))) for x in m_x]
    return curve_y
    
rmse_my_poly_train_error = rmse(beta_my_polynom_train, xTrain, yTrain)
rmse_my_ridge_train_error = rmse(beta_my_ridge_train, xTrain, yTrain)
rmse_my_ridge_poly_train_error = rmse(beta_my_ridge_polynom_test, xTrain, yTrain)
print('rmse_my_poly_train_error: ', rmse_my_poly_train_error)
print('rmse_my_ridge_train_error: ', rmse_my_ridge_train_error)
print('rmse_my_ridge_poly_train_error: ', rmse_my_ridge_poly_train_error)

#FOR TEST
rmse_my_poly_test_error = rmse(beta_my_polynom_test, xTest, yTest)
rmse_my_ridge_test_error = rmse(beta_my_ridge_train, xTest, yTest)
rmse_my_ridge_poly_test_error = rmse(beta_my_ridge_polynom_test, xTest, yTest)
print('rmse_my_poly_test_error: ', rmse_my_poly_test_error)
print('rmse_my_ridge_test_error: ', rmse_my_ridge_test_error)
print('rmse_my_ridge_poly_test_error: ', rmse_my_ridge_poly_test_error)
```

```python
def plot_fit_line(x, y, curve_x, curve_y_arr, xlabel, ylabel, title, labels):
    plt.figure(figsize=(15,10))
    plt.scatter(x, y, s=4, color='blue')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    for i in range(len(curve_y_arr)):
        color = np.random.rand(3)
        plt.plot(curve_x, curve_y_arr[i], color=color, label=labels[i], linewidth=3, zorder=10-i)
        plt.legend(loc='best')
    plt.show()   
```
